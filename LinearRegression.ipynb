    <ul style="font-size: 32px; line-height: 1.8; list-style: none; padding: 0; margin-top: 40px;">
        <li style="margin-bottom: 30px;">
            • A <span style="color: #3274B5;">linear function</span> of features \(\textit{x}_1, \textit{x}_2, \ldots, \textit{x}_D\) is a function that takes the form:
            \[
                f_{\mathbf{w}}(x_1, x_2, \ldots, x_D) = w_0 + w_1x_1 + \ldots + w_Dx_D = \sum_{d=1}^{D} w_d x_d = \mathbf{w}^T \cdot \mathbf{x}

        </li>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px;">Linear Regression</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="display: flex; flex-direction: row;">
        <div style="width: 55%;">
            <ul style="font-size: 28px; list-style: none; margin-top: 15px;">
                <li>
                    • <it>“Linear regression is the problem of fitting a </it><span style="color: #3274B5;">linear function</span> <it>to a set of \(n\) examples, in which <span style="color: #3274B5;">usually<span> the input and output features are <span style="color: #3274B5;">numeric</span></it>”.
                </li>
                <li>
                    • Aim: <span style="color: #A62B17;">Predict</span> the <span style="color: #3274B5;">output feature</span> \(\hat{Y}\) from input features \(\textit{x}_1, \textit{x}_2, \ldots, \textit{x}_D\)
                </li>
                <li>
                    • Notation:
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li> 
                           ▸ A <span style="color: #3274B5;">feature</span> is a <span style="color: #A62B17;">function</span> of an example \(\rightarrow\) often <span style="color: #3274B5;">human interpretation</span>
                        </li>
                        <li> 
                           ▸ The <span style="color: #3274B5;">Weights</span> denote the "<span style="color: #A62B17;">importance</span>" of features
                        </li>
                        <li> 
                           ▸ <span style="color: #3274B5;">sum of squared error (SSE)</span> on examples \(X\) for output features \(Y\) is:
                            \[
                                SSE(\mathbf{X}, \mathbf{w}) = \sum_{\mathbf{x}_n \in X} (Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n))^2
                            \]
                        </li>
                    </ul>
                </li>
                <li>
                    • Goal: Find <span style="color: #3274B5;">weights \(x\)</span> that <span style="color: #A62B17;">minimise \( SSE(X, w) \ldots \)</span> 
                </li>
                <li>
                    • Identical to optimising a <span style="color: #3274B5;">linear model</span> wrt the <span style="color: #3274B5;">squared objective function</span> \(l^{(sqr)}\) over \(n\) examples <span style="color: #3274B5;">without regularisation:</span>
                    \[
                        \min_{\mathbf{w}, b} \sum (y_n - (\mathbf{w}^T \cdot \mathbf{x}_n + b))^2, \text{ where } w_0 = b \text{ (bias term) and } x_0 = 1
                    \]
                </li>
                <li>
                    • The gradient becomes \(\nabla_{\mathbf{w}} = (y_n - \mathbf{w}^T \cdot \mathbf{x}_n)\mathbf{x}_n\) 
                </li>
            </ul>
        </div>
        <div style="width: 30%; margin-top: 1%; margin-left: 15%;">

        </div>
    </div>

        &nbsp;
    </div>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">Linear Regression: Gradient Descent Algorithm</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="width: 70%; font-size: 20px;">
    \[
        \begin{array}{l}
        \hline
        \color{purple}{\textbf{Algorithm: }} \text{LINEARREGRESSIONGRADIENTDESCENT}(\text{data}, maxIter, \eta) \\
        \hline
        \textbf{1: } \mathbf{w} \leftarrow \langle 0,0,\ldots,0 \rangle \text{ or randomly initialise} \quad \textit{// initialise weights} \\
        \textbf{2: } \textbf{for } k = 1 \ldots maxIter \textbf{ do} \\
        \textbf{3: } \quad \textbf{for all } (x, y) \in \text{data} \textbf{ do} \\
        \textbf{4: } \qquad \hat{Y}(\mathbf{x}_n) = \mathbf{w}^T \cdot \mathbf{x}_n \quad \textit{// compute predicted output} \\
        \textbf{5: } \qquad \delta = Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n) \quad \textit{// compute scalar } \delta \\
        \textbf{6: } \qquad \mathbf{w} \leftarrow \mathbf{w} + \eta \cdot \delta \cdot \mathbf{x}_n \quad \textit{// update weights} \\
        \textbf{7: } \quad \textbf{end for} \\
        \textbf{8: } \textbf{end for} \\
        \textbf{9: } \textbf{return } \mathbf{w} \\
        \hline
        \end{array}
    \]

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pandas as pd

# Data
data = pd.DataFrame({
    'Music DL?': [0, 1, 1, 1, 0, 0, 0, 0, 1, 1],
    'Music Stream?': [1, 0, 0, 0, 1, 1, 0, 1, 1, 1],
    'Online Games': [1, 0, 1, 0, 0, 0, 1, 1, 0, 1],
    'Output': [1, 0, 1, 0, 0, 0, 1, 0, 0, 1]
}, index=[f'T{i}' for i in range(1, 11)])

print("Training Data:")
display(data)

# Add bias term (the "1" at the beginning)
X = data.iloc[:, :-1].values
X_with_bias = np.column_stack([np.ones(len(X)), X])  # Add column of 1s
y = data['Output'].values

print("\nX with bias term (w_0)):")
print(X_with_bias[:2])
print(f"\nT1: x = {X_with_bias[0]}, y = {y[0]}")
print(f"T2: x = {X_with_bias[1]}, y = {y[1]}")
```

```python
# Walkthrough of first iteration
print("=" * 60)
print("Let's train on first 2 examples")
print("=" * 60)

# Initialize weights randomly
w = np.array([1, 2, 3, 1])  # [w0, w1, w2, w3]
eta = 1  # learning rate

print(f"\nInitial weights: w = {w}")
print(f"Learning rate: η = {eta}")

# Iteration 1, Example T1
print("\n" + "="*60)
print("ITERATION 1 - Example T1")
print("="*60)
x_T1 = X_with_bias[0]  # [1, 0, 1, 1]
y_T1 = y[0]  # 1

print(f"\nx_T1 = {x_T1}")
print(f"y_T1 = {y_T1}")

# Line 4: Compute prediction
y_hat = np.dot(w, x_T1)
print(f"\nŷ(x_T1) = w^T · x_T1")
print(f"        = {w} · {x_T1}")
print(f"        = {w[0]}*{x_T1[0]} + {w[1]}*{x_T1[1]} + {w[2]}*{x_T1[2]} + {w[3]}*{x_T1[3]}")
print(f"        = {y_hat}")

# Line 5: Compute error
delta = y_T1 - y_hat
print(f"\nδ = Y(x_T1) - ŷ(x_T1)")
print(f"  = {y_T1} - {y_hat}")
print(f"  = {delta}")

# Line 6: Update weights
w_old = w.copy()
w = w + eta * delta * x_T1
print(f"\nw^(1) = w^(0) + η · δ · x_T1")
print(f"      = {w_old} + {eta} * {delta} * {x_T1}")
print(f"      = {w_old} + {eta * delta * x_T1}")
print(f"      = {w}")
```

```python
# Iteration 1, Example T2
print("\n" + "="*60)
print("ITERATION 1 - Example T2")
print("="*60)
x_T2 = X_with_bias[1] # [1, 0, 0]
y_T2 = y[1]  # 0

print(f"\nx_T2 = {x_T2}")
print(f"y_T2 = {y_T2}")

# Line 4: Compute prediction
y_hat = np.dot(w, x_T2)
print(f"\nŷ(x_T2) = w^T · x_T2")
print(f"        = {w} · {x_T2}")
print(f"        = {w[0]}*{x_T2[0]} + {w[1]}*{x_T2[1]} + {w[2]}*{x_T2[2]} + {w[3]}*{x_T2[3]}")
print(f"        = {y_hat}")

# Line 5: Compute error
delta = y_T2 - y_hat
print(f"\nδ = Y(x_T2) - ŷ(x_T2)")
print(f"  = {y_T2} - {y_hat}")
print(f"  = {delta}")

# Line 6: Update weights
w_old = w.copy()
w = w + eta * delta * x_T2
print(f"\nw^(1) = w^(0) + η · δ · x_T2")
print(f"      = {w_old} + {eta} * {delta} * {x_T2}")
print(f"      = {w_old} + {eta * delta * x_T2}")
print(f"      = {w}")
```

```python
# After iteration 1, let's make predictions with the updated weights
print("\n" + "="*60)
print("MAKING PREDICTIONS AFTER ITERATION 1")
print("="*60)

print(f"\nWeights after iteration 1: w = {w}")
print(f"\nLet's predict on the first 2 examples (outside the training loop)")

# Predict on T1
print("\n" + "-"*60)
print("Prediction for T1:")
print("-"*60)
x_T1 = X_with_bias[0]  # [1, 0, 1, 1]
y_T1 = y[0]  # 1

y_hat_T1 = np.dot(w, x_T1)
print(f"\nŷ(x_T1) = w^T · x_T1")
print(f"        = {w} · {x_T1}")
print(f"        = {w[0]:.1f}*{x_T1[0]} + {w[1]:.1f}*{x_T1[1]} + {w[2]:.1f}*{x_T1[2]} + {w[3]:.1f}*{x_T1[3]}")
print(f"        = {y_hat_T1:.1f}")
print(f"\nActual: y_T1 = {y_T1}")
print(f"Error: {y_T1} - {y_hat_T1:.1f} = {y_T1 - y_hat_T1:.1f}")

# Predict on T2
print("\n" + "-"*60)
print("Prediction for T2:")
print("-"*60)
x_T2 = X_with_bias[1]  # [1, 1, 0, 0]
y_T2 = y[1]  # 0

y_hat_T2 = np.dot(w, x_T2)
print(f"\nŷ(x_T2) = w^T · x_T2")
print(f"        = {w} · {x_T2}")
print(f"        = {w[0]:.1f}*{x_T2[0]} + {w[1]:.1f}*{x_T2[1]} + {w[2]:.1f}*{x_T2[2]} + {w[3]:.1f}*{x_T2[3]}")
print(f"        = {y_hat_T2:.1f}")
print(f"\nActual: y_T2 = {y_T2}")
print(f"Error: {y_T2} - {y_hat_T2:.1f} = {y_T2 - y_hat_T2:.1f}")

# Calculate SSE for these two examples
print("\n" + "="*60)
print("CALCULATING ERROR (SSE)")
print("="*60)
sse = (y_T1 - y_hat_T1)**2 + (y_T2 - y_hat_T2)**2
print(f"\nSSE(X, w) = Σ (Y(x_n) - ŷ(x_n))²")
print(f"          = (Y(x_T1) - ŷ(x_T1))² + (Y(x_T2) - ŷ(x_T2))²")
print(f"          = ({y_T1} - ({y_hat_T1:.1f}))² + ({y_T2} - ({y_hat_T2:.1f}))²")
print(f"          = ({y_T1 - y_hat_T1:.1f})² + ({y_T2 - y_hat_T2:.1f})²")
print(f"          = {(y_T1 - y_hat_T1)**2:.1f} + {(y_T2 - y_hat_T2)**2:.1f}")
print(f"          = {sse:.1f}")

print("\n" + "="*60)
print("Obviously, the predictions are way off... why?")
print("Because we've only done ONE iteration of training!")
print("The model needs more iterations to learn the patterns.")
print("="*60)
```

```python
# Full training process with visualization
def train_with_history(X, y, eta=0.3, max_iter=10):
    """Train and record history for visualization"""
    n_samples, n_features = X.shape
    w = np.array([1, 2, 3, 1], dtype=float)  # Initialize
    
    history = {
        'weights': [w.copy()],
        'predictions': [],
        'errors': [],
        'iteration': [0],
        'example': ['Initial']
    }
    
    for iteration in range(1, max_iter + 1):
        for i in range(n_samples):
            # Prediction
            y_hat = np.dot(w, X[i])
            
            # Error
            delta = y[i] - y_hat
            
            # Update
            w = w + eta * delta * X[i]
            
            # Record
            history['weights'].append(w.copy())
            history['predictions'].append(y_hat)
            history['errors'].append(delta)
            history['iteration'].append(iteration)
            history['example'].append(f'T{i+1}')
    
    return w, history

# Train
final_w, history = train_with_history(X_with_bias, y, eta=0.3, max_iter=5)

print(f"Final weights after 5 iterations: {final_w}")
```

```python
# Loss over time
fig, ax1 = plt.subplots(1, 1, figsize=(14, 5))

# Calculate SSE (Sum of Squared Errors) over iterations
losses = []
for step in range(0, len(history['weights']), len(X_with_bias)):
    w_step = history['weights'][step]
    y_pred = np.dot(X_with_bias, w_step)  # Raw predictions
    sse = np.sum((y - y_pred) ** 2)  # Sum of squared errors
    losses.append(sse)

ax1.plot(range(len(losses)), losses, 'o-', linewidth=2, markersize=8, color='red')
ax1.set_xlabel('Iteration', fontsize=12)
ax1.set_ylabel('SSE (Sum of Squared Errors)', fontsize=12)
ax1.set_title('Training Loss over Iterations', fontsize=13)
ax1.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Final predictions
y_pred_final = np.sign(np.dot(X_with_bias, final_w))
y_pred_final[y_pred_final == 0] = 1

comparison = pd.DataFrame({
    'True Output': y,
    'Predicted': y_pred_final.astype(int),
    'Correct': y == y_pred_final
}, index=[f'T{i}' for i in range(1, 11)])

print("\nFinal Comparison:")
print(comparison)
print(f"\nFinal Accuracy: {np.mean(y == y_pred_final):.1%}")
print(f"Final SSE: {losses[-1]:.2f}")
```

    <ul style="font-size: 20px; list-style: none; margin-top: 15px;">
                <li style="margin-top: 20px;">
                    • Recall our optimisation problem has become: \(L(w) = \textit{min}_w \sum_n (y_n - (w^T \cdot x_n))^2\)
                </li>
                <li style="margin-top: 20px;">
                    • Converting to matrix form \(w^T \cdot x_n = x_{n}^{T} \cdot w = X \cdot w\), and \(L(w)\) becomes \((Y - X \cdot w)^T (Y - X \cdot w)\), where
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ \(Y\) is the vector of all outputs \((N \times 1)\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(X\) is the matrix of all inputs \((N \times D)\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(w\) is the weight vector \((D \times 1)\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \((Y - X \cdot w)^T (Y - X \cdot w)\) is the dot product of the error vector by itself (to get the sum of squared errors)
                        </li>
                    </ul>
                </li>
                <li style="margin-top: 20px;">
                    • Expanding \(L(w) = (Y - X \cdot w)^T (Y - X \cdot w)\)
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ \(= Y^T Y - Y^T(Xw) - (Xw)^T Y + (Xw)^T(Xw)\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(= Y^T Y - Y^T Xw - (Xw)^T Y + (Xw)^T Xw\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(= Y^T Y - Y^T Xw - w^T X^T Y + w^T X^T Xw\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(= Y^T Y - 2w^T X^T Y + w^T X^T Xw\)
                            <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                                <li style="margin-top: 15px;">
                                    - as \(Y^T Xw = (Y^T Xw)^T = w^T X^T Y\); meaning that \(Y^T Xw\) and \(w^T X^T Y\) are the same
                                </li>
                            </ul>
                        </li>
                    </ul>
                </li>
            </ul>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">Linear Regression: Closed Form Solution</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <ul style="font-size: 20px; list-style: none; margin-top: 15px;">
                <li style="margin-top: 20px;">
                    • Deriving \(L(w) = Y^T Y - 2w^T X^T Y + w^T X^T Xw\)
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ \(\frac{\partial}{\partial{w}}(Y^T \cdot Y) = 0\); 
                            <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                                <li style="margin-top: 15px;">
                                    - because \((Y^T \cdot Y)\) is a constant
                                </li>
                            </ul>
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(\frac{\partial}{\partial{w}} (-2w^T X^T Y) = -2(X^T Y)\); 
                            <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                                <li style="margin-top: 15px;">
                                    - as \(\frac{\partial}{\partial{w}} (w^T a) = a\) where \(a\) is a constant and \(a = X^T Y\) here
                                </li>
                            </ul>
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(\frac{\partial}{\partial{w}} (w^T X^T X w) = 2(X^T X)w\);
                            <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                                <li style="margin-top: 15px;">
                                    - as \(\frac{\partial}{\partial{w}} (w^T W w) = 2Aw\) if \(A\) is symmetric and \(A = X^T X\) here
                                </li>
                            </ul>
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ Therefore: \(\Delta_w L(w) = 0 - 2X^T Y + 2X^T Xw\) </br>
                            \(\Delta_w L(w) = 2X^T Xw - 2X^T Y\)
                        </li>
                    </ul>
                </li>
                <li style="margin-top: 20px;">
                    • By <span style="color: #A62B17;">setting the derivative of \(\Delta_w L(W) = 0\)</span> we can arrive at:
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ \(2X^T Xw - 2X^T Y = 0\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(X^T Xw - X^T Y = 0\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(X^T Xw = X^T Y\)
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(w = \frac{(X^T Y)}{(X^T X)}\)
                        </li>
                        <li> 
                           ▸ <span style="color: #3274B5;">\(w = (X^TX)^{-1}X^TY\)</span> where \(X\) is the dataset \((N \times D)\), \(Y\) are all the gold output values \((N \times 1)\)
                        </li>
                    </ul>
                </li>
            </ul>

    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <ul style="font-size: 20px; list-style: none; margin-top: 15px;">
                <li style="margin-top: 20px;">
                    • Why not do this then?
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ <span style="color: #A62B17;">Complexity</span> of <span style="color: #3274B5;">Gradient Descent</span> is <span style="color: #3274B5;">\(\mathcal{O}(KND)\)</span>, where \(K\) are the number of iterations
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ <span style="color: #A62B17;">Complexity</span> of <span style="color: #3274B5;">Closed Form Solution</span> is <span style="color: #3274B5;">\(\mathcal{O}(D^3 + D^2N)\)</span>
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(\frac{\partial}{\partial{w}} (w^T X^T X w) = 2(X^T X)w\);
                            <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                                <li style="margin-top: 15px;">
                                    - Most of the times \(N > D\), hence complexity is dominated by <span style="color: #3274B5;">\(\mathcal{O}(D^2N)\)</span>
                                </li>
                            </ul>
                        </li>
                    </ul>
                </li>
                <li>
                    • Do we need to run more than \(D\)-many iterations \(K\)???
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ If we can get good results with \(K \lt D\) (or even \(K \lt\lt D)\), then gradient descent can be very efficient, especially for high-dimensional problems where \(D^3\) is prohibitively expensive.
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ In practice, we often don't need \(K \lt\lt D\), which is one reason gradient descent is widely used in machine learning
                        </li>
                    </ul>
                </li>
                <li>
                    • Rule of thumb:
                    <ul style="list-style: none; margin-left: 20px; margin-top: 15px;">
                        <li style="margin-top: 15px;"> 
                           ▸ \(D \le 100 \rightarrow\) closed-form is faster
                        </li>
                        <li style="margin-top: 15px;"> 
                           ▸ \(D \ge 10,000 \rightarrow\) probably gradient descent is faster
                        </li>
                    </ul>
                </li>
            </ul>

```python
import time

# Demonstrate both methods on small data
print("="*70)
print("EXAMPLE: Comparing Both Methods")
print("="*70)

# Use our existing data
X_demo = X_with_bias
y_demo = y
N, D = X_demo.shape

print(f"\nDataset: N={N} examples, D={D} features (including bias)")

# Method 1: Gradient Descent (we already did this)
start_time = time.time()
w_gd, _ = train_with_history(X_demo, y_demo, eta=0.01, max_iter=100)
time_gd = time.time() - start_time

# Method 2: Closed Form Solution
start_time = time.time()
w_closed = np.linalg.inv(X_demo.T @ X_demo) @ X_demo.T @ y_demo
time_closed = time.time() - start_time

print(f"\n{'Method':<20} {'Time (ms)':<15} {'Weights'}")
print("-"*70)
print(f"{'Gradient Descent':<20} {time_gd*1000:>10.4f} ms   {w_gd}")
print(f"{'Closed Form':<20} {time_closed*1000:>10.4f} ms   {w_closed}")

# Compare predictions
y_pred_gd = X_demo @ w_gd
y_pred_closed = X_demo @ w_closed
sse_gd = np.sum((y_demo - y_pred_gd)**2)
sse_closed = np.sum((y_demo - y_pred_closed)**2)

print(f"\n{'Method':<20} {'SSE (Error)':<15}")
print("-"*70)
print(f"{'Gradient Descent':<20} {sse_gd:>10.4f}")
print(f"{'Closed Form':<20} {sse_closed:>10.4f}")
print(f"\n→ Closed form gives the EXACT optimal solution!")
print(f"→ Gradient descent approximates it (depends on iterations)")
```

```python
# Visualize the complexity comparison
print("\n" + "="*70)
print("COMPUTATIONAL COMPLEXITY: When to use which method?")
print("="*70)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Scenario 1: Varying number of features (D), fixed N=1000
N_fixed = 1000
D_range = np.arange(10, 500, 10)
K = 100  # iterations for GD

time_gd_D = K * N_fixed * D_range  # O(K*N*D)
time_closed_D = D_range**3 + D_range**2 * N_fixed  # O(D³ + D²N)

ax1.plot(D_range, time_gd_D / 1e6, label=f'Gradient Descent (K={K} iterations)', 
         linewidth=2, color='blue')
ax1.plot(D_range, time_closed_D / 1e6, label='Closed Form', 
         linewidth=2, color='red')
ax1.axvline(x=100, color='green', linestyle='--', alpha=0.7, linewidth=2)
ax1.text(105, max(time_closed_D / 1e6) * 0.8, 'D=100\nRule of thumb', 
         fontsize=11, color='green', weight='bold')
ax1.set_xlabel('Number of Features (D)', fontsize=13)
ax1.set_ylabel('Relative Computation (millions of ops)', fontsize=13)
ax1.set_title(f'Fixed N={N_fixed} samples', fontsize=14, weight='bold')
ax1.legend(fontsize=11)
ax1.grid(alpha=0.3)

# Scenario 2: Varying number of samples (N), fixed D=50
D_fixed = 50
N_range = np.arange(100, 20000, 500)

time_gd_N = K * N_range * D_fixed  # O(K*N*D)
time_closed_N = D_fixed**3 + D_fixed**2 * N_range  # O(D³ + D²N)

ax2.plot(N_range, time_gd_N / 1e6, label=f'Gradient Descent (K={K} iterations)', 
         linewidth=2, color='blue')
ax2.plot(N_range, time_closed_N / 1e6, label='Closed Form', 
         linewidth=2, color='red')
ax2.axvline(x=10000, color='green', linestyle='--', alpha=0.7, linewidth=2)
ax2.text(10500, max(time_gd_N / 1e6) * 0.7, 'N=10,000\nRule of thumb', 
         fontsize=11, color='green', weight='bold')
ax2.set_xlabel('Number of Samples (N)', fontsize=13)
ax2.set_ylabel('Relative Computation (millions of ops)', fontsize=13)
ax2.set_title(f'Fixed D={D_fixed} features', fontsize=14, weight='bold')
ax2.legend(fontsize=11)
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">Use Case: Gradient Calculation for \(l^{(sqr)}\)</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="display: flex; flex-direction: row; justify-content: center;">
        <ul style="font-size: 28px; line-height: 1.8; list-style: none; padding: 0; margin-top: 40px;">
            <li style="margin-bottom: 20px;">
                • Consider the following <span style="color: #3274B5;">learning objective</span>
                \[
                    L(\mathbf{w}) = \sum_{n} \left( Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n) \right)^2, \text{ where } \hat{Y}(\mathbf{x}_n) = \sigma(\mathbf{w}^T \cdot \mathbf{x}_n)
                \]
            </li>
            <li style="margin-bottom: 20px;">
                • What do we need in order to run <span style="color: #3274B5;">gradient descent</span>?
            </li>
        </ul>
    </div>
    <div style="position: absolute; bottom: 20px; left: 50%; transform: translateX(-50%); font-size: 24px; color: #666;">10</div>
    <div style="display: flex; flex-direction: column; align-items: center; justify-content: center; height: 60px; text-align: center;">&nbsp;</div>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">\(l^{(sqr)}\): Derivative with respect to w</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="text-align: left; margin-left: 0; font-size: 18px;">
                \[
\nabla_{\mathbf{w}} L = \nabla_{\mathbf{w}} \sum_{n} \left( Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n) \right)^2
\]
                \[
                    = \nabla_{\mathbf{w}} \sum_{n} \left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right)^2
                \]
                // power rule (brings down the 2) + chain rule (multiply by derivative of inside)
                \[
= \sum_{n} 2\left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) \nabla_{\mathbf{w}} \left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right)
                \]
                // sum rule: the derivative of a difference is the difference of derivatives
                \[
= \sum_{n} 2\left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) -\nabla_{\mathbf{w}} \sigma(\mathbf{w}^T \cdot \mathbf{x}_n)
                \]
                // chain Rule: the derivative of the sigmoid function is sigma times one minus sigma.
                \[
= \sum_{n} -2\left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \left( 1 - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) \nabla_{\mathbf{w}} \mathbf{w}^T \cdot \mathbf{x}_n
                \]
                // constant rule: the derivative of the dot product w^T·x with respect to w is x
                \[
= \sum_{n} -2\left( Y(\mathbf{x}_n) - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \left( 1 - \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \right) \mathbf{x}_n
                \]
                // since Y^(x_n)=σ(w^T⋅x_n)
                \[
                    \sum_n -2(Y(x_n) - \hat{Y}(x_n)) \hat{Y}(x_n) (1 - \hat{Y}(x_n)) x_n
                \]
                // let the error (delta) =Y(x_n)−Y^(x)n)
                \[
= \sum_{n} -2 \delta \hat{Y}(\mathbf{x}_n) \left( 1 - \hat{Y}(\mathbf{x}_n) \right) \cdot x_n, \quad \text{where } \delta = Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n)
                \]
    </div>
    <div style="position: absolute; bottom: 20px; left: 50%; transform: translateX(-50%); font-size: 24px; color: #666;">10</div>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">Logistic Regression: Gradient Descent Algorithm</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="width: 70%; font-size: 18px;">
    \[
        \begin{array}{l}
        \hline
        \color{purple}{\textbf{Algorithm: }} \text{LOGISTICREGRESSIONGRADIENTDESCENT}(\text{data}, maxIter, \eta) \\
        \hline
        \textbf{1: } \mathbf{w} \leftarrow \langle 0,0,\ldots,0 \rangle \text{ or randomly initialise} \quad \textit{// initialise weights} \\
        \textbf{2: } \textbf{for } k = 1 \ldots maxIter \textbf{ do} \\
        \textbf{3: } \quad \textbf{for all } (x, y) \in \text{data} \textbf{ do} \\
        \textbf{4: } \quad \quad \hat{Y}(\mathbf{x}_n) = \sigma(\mathbf{w}^T \cdot \mathbf{x}_n) \quad \textit{// compute predicted output} \\
        \textbf{5: } \quad \quad \delta = Y(\mathbf{x}_n) - \hat{Y}(\mathbf{x}_n) \quad \textit{// compute scalar } \delta \\
        \textbf{6: } \quad \quad \mathbf{w} \leftarrow \mathbf{w} + \eta \cdot \delta \cdot \hat{Y}(\mathbf{x}_n)(1 - \hat{Y}(\mathbf{x}_n)) \cdot \mathbf{x}_n \quad \textit{// update weights} \\
        \textbf{7: } \quad \textbf{end for} \\
        \textbf{8: } \textbf{end for} \\
        \textbf{9: } \textbf{return } \mathbf{w} \\
        \hline
        \end{array}
    \]

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">\(l^{(log)}\): Log Loss for Logistic Regression</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="display: flex; flex-direction: row; justify-content: center;">
        <ul style="font-size: 28px; line-height: 1.8; list-style: none; padding: 0; margin-top: 40px;">
            <li style="margin-bottom: 20px;">
                • Quite often we optimise instead of the squared loss \(l^{(sqr)}\), the log-loss \(l^{(log)}\)
                \[
                    \begin{align}
                    LL(\mathbf{w}) &= -\sum_{n} Y(\mathbf{x}_n) \cdot \log \hat{Y}(\mathbf{x}_n) + (1 - Y(\mathbf{x}_n)) \cdot \log(1 - \hat{Y}(\mathbf{x}_n)) \\
                    \text{where } \hat{Y}(\mathbf{x}_n) &= \sigma(\mathbf{w}^T \cdot \mathbf{x}_n)
                    \end{align}
                \]
            </li>
            <li style="margin-bottom: 20px;">
                • Log-loss is the natural choice for logistic regression because it comes from probability theory and has much better optimization properties (better gradients and faster learning)
            </li>
        </ul>
    </div>

    <h1 style="font-size: 72px; font-weight: 700; margin-left: 10px; color: #333;">\(l^{(log)}\): Log Loss for Logistic Regression</h1>
    <hr style="width: 50%; margin-left: -0%; margin-top: -0%;" />
    <div style="font-size: 18px;">
        <h2>STEP 1: DERIVATIVE OF LOG-LOSS</h2>
            For a single example $n$, we have:    
            $$LL_n = -Y(\mathbf{x}_n) \cdot \log \hat{Y}(\mathbf{x}_n) - (1-Y(\mathbf{x}_n)) \cdot \log(1-\hat{Y}(\mathbf{x}_n))$$
            Let's compute $\frac{\partial LL_n}{\partial \mathbf{w}}$ using the chain rule:
            $$\frac{\partial LL_n}{\partial \mathbf{w}} = \frac{\partial LL_n}{\partial \hat{Y}} \cdot \frac{\partial \hat{Y}}{\partial \mathbf{w}}$$
            ---
            <b>Part 1:</b> $\frac{\partial LL_n}{\partial \hat{Y}}$        
            $$\frac{\partial}{\partial \hat{Y}} \left[-Y \cdot \log(\hat{Y}) - (1-Y) \cdot \log(1-\hat{Y})\right]$$
            $$= -Y \cdot \frac{1}{\hat{Y}} - (1-Y) \cdot \frac{1}{1-\hat{Y}} \cdot (-1)$$
            $$= -\frac{Y}{\hat{Y}} + \frac{1-Y}{1-\hat{Y}}$$
            $$= \frac{-Y(1-\hat{Y}) + (1-Y)\hat{Y}}{\hat{Y}(1-\hat{Y})}$$
            $$= \frac{-Y + Y\hat{Y} + \hat{Y} - Y\hat{Y}}{\hat{Y}(1-\hat{Y})}$$
            $$= \frac{\hat{Y} - Y}{\hat{Y}(1-\hat{Y})}$$
            ---
            <b>Part 2:</b> $\frac{\partial \hat{Y}}{\partial \mathbf{w}}$ where $\hat{Y} = \sigma(\mathbf{w}^T \cdot \mathbf{x})$
            We need the derivative of the sigmoid:
            $$\frac{\partial \sigma(z)}{\partial z} = \sigma(z)(1 - \sigma(z))$$
            So by chain rule:
            $$\frac{\partial \hat{Y}}{\partial \mathbf{w}} = \frac{\partial \sigma(\mathbf{w}^T \cdot \mathbf{x})}{\partial \mathbf{w}}$$
            $$= \sigma(\mathbf{w}^T \cdot \mathbf{x})(1-\sigma(\mathbf{w}^T \cdot \mathbf{x})) \cdot \frac{\partial (\mathbf{w}^T \cdot \mathbf{x})}{\partial \mathbf{w}}$$
            $$= \hat{Y}(1-\hat{Y}) \cdot \mathbf{x}$$
            ---
            <b>Part 3:</b> Combine them
            $$\frac{\partial LL_n}{\partial \mathbf{w}} = \frac{\partial LL_n}{\partial \hat{Y}} \cdot \frac{\partial \hat{Y}}{\partial \mathbf{w}}$$
            $$= \frac{\hat{Y} - Y}{\hat{Y}(1-\hat{Y})} \cdot \hat{Y}(1-\hat{Y}) \cdot \mathbf{x}$$
            $$= (\hat{Y} - Y) \cdot \mathbf{x} \quad$$
            $$= -(Y - \hat{Y}) \cdot \mathbf{x}$$
            $$= -\delta \cdot \mathbf{x} \quad \text{where } \delta = Y - \hat{Y}$$
            <div style="background-color: #ffffcc; padding: 15px; border-left: 5px solid #ffcc00; margin: 20px 0;">
            The sigmoid derivative $\hat{Y}(1-\hat{Y})$ perfectly cancels with the denominator from the log derivative! This is why log-loss works well with sigmoid.
            </div>

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    'Thread': ['old', 'new', 'new'],
    'Length': ['long', 'short', 'long'],
    'Where': ['work', 'work', 'home'],
    'Action': ['skip', 'skip', 'read']
}, index=['E1', 'E2', 'E3'])

df.index.name = 'Ex'

# Display the DataFrame
display(df)
```

```python
import pandas as pd

# Create the vectorized DataFrame
df_vectorized = pd.DataFrame({
    'Thread': [0, 1, 1],
    'Length': [0, 1, 0],
    'Where': [0, 0, 1],
    'Output': [0, 0, 1]
}, index=['E1', 'E2', 'E3'])

df_vectorized.index.name = 'Ex'

display(df_vectorized)
```

```python
import pandas as pd
import numpy as np

# Add bias term (w0)
X = df_vectorized[['Thread', 'Length', 'Where']].values
X_with_bias = np.column_stack([np.ones(len(X)), X])
y = df_vectorized['Output'].values

print("\n" + "="*70)
print("LEARNING THE FUNCTION")
print("="*70)
print("\nWe want to learn:")
print("  f(thread, length, where) = σ(w₀ + w₁·thread + w₂·length + w₃·where)")
print("\nMinimise SSE (squared loss) or log-loss")
print()
```

```python
# Initialize weights
w = np.array([0.4, -0.7, -0.9, 1.2])
print(f"Initialize weights: w = {w}")
print(f"  w₀ (bias) = {w[0]}")
print(f"  w₁ (Thread) = {w[1]}")
print(f"  w₂ (Length) = {w[2]}")
print(f"  w₃ (Where) = {w[3]}")
```

```python
print("\n" + "="*70)
print("MAKING PREDICTIONS")
print("="*70)
print("\nHow to predict class based on output:")
print("  If σ(w^T·x) ≥ 0.5 → predict 1 (read)")
print("  If σ(w^T·x) < 0.5 → predict 0 (skip)")
print()

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Make predictions
results = []
for i, ex in enumerate(['E1', 'E2', 'E3']):
    x = X_with_bias[i]
    raw_score = np.dot(w, x)
    prob = sigmoid(raw_score)
    predicted = 1 if prob >= 0.5 else 0
    gold = y[i]
    delta = gold - prob
    sse = delta ** 2
    
    results.append({
        'Ex': ex,
        'Thread': X[i, 0],
        'Length': X[i, 1],
        'Where': X[i, 2],
        'Output (Predicted / Gold)': f'f({raw_score:.1f}) = {prob:.2f} / {gold}',
        'δ': f'{delta:.2f}',
        'SSE': f'{sse:.3f}'
    })

df_results = pd.DataFrame(results).set_index('Ex')
display(df_results)
```

```python
print("\n" + "="*70)
print("DETAILED CALCULATIONS")
print("="*70)

for i, ex in enumerate(['E1', 'E2', 'E3']):
    x = X_with_bias[i]
    raw_score = np.dot(w, x)
    prob = sigmoid(raw_score)
    predicted = 1 if prob >= 0.5 else 0
    gold = y[i]
    delta = gold - prob
    sse = delta ** 2
    
    print(f"\n{ex}: x = [1, {X[i,0]}, {X[i,1]}, {X[i,2]}] (with bias)")
    print(f"  Raw score: w^T·x = {w[0]}×1 + {w[1]}×{X[i,0]} + {w[2]}×{X[i,1]} + {w[3]}×{X[i,2]}")
    print(f"            = {raw_score:.2f}")
    print(f"  Probability: σ({raw_score:.2f}) = {prob:.3f}")
    print(f"  Predicted class: {predicted} ({'read' if predicted == 1 else 'skip'})")
    print(f"  True class: {gold} ({'read' if gold == 1 else 'skip'})")
    print(f"  Error (δ): {gold} - {prob:.3f} = {delta:.3f}")
    print(f"  Squared error: ({delta:.3f})² = {sse:.4f}")

total_sse = np.sum((y - sigmoid(X_with_bias @ w))**2)
print(f"\n{'='*70}")
print(f"Total SSE = {total_sse:.4f}")
print(f"{'='*70}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
import matplotlib.pyplot as plt

print("="*70)
print("COMPARING LINEAR VS LOGISTIC REGRESSION")
print("="*70)

data = pd.DataFrame({
    'Music DL?': [0, 1, 1, 1, 0, 0, 0, 0, 1, 1],
    'Music Stream?': [1, 0, 0, 0, 1, 1, 0, 1, 1, 1],
    'Online Games': [1, 0, 1, 0, 0, 0, 1, 1, 0, 1],
    'Output': [1, 0, 1, 0, 0, 0, 1, 0, 0, 1]
}, index=[f'T{i}' for i in range(1, 11)])

print("\nTraining Data:")
display(data)

X = data.iloc[:, :-1].values
y = data['Output'].values
```

```python
print("\n" + "="*70)
print("METHOD 1: LINEAR REGRESSION")
print("="*70)

# Train linear regression
lr = LinearRegression()
lr.fit(X, y)

# Get predictions
y_pred_linear = lr.predict(X)

# Create results table
results_linear = pd.DataFrame({
    'Transaction': [f'T{i}' for i in range(1, 11)],
    'Actual': y,
    'Predicted': y_pred_linear.round(3),
    'Error': (y - y_pred_linear).round(3)
})

print("\nLinear Regression Results:")
display(results_linear)

print("\n" + "="*70)
print("LEARNED FUNCTION (Linear Regression):")
print("="*70)
print(f"\nBuys = f(musicDown, musicStream, onlineGames)")
print(f"     = {lr.intercept_:.3f}")
print(f"       {lr.coef_[0]:+.3f} * musicDown")
print(f"       {lr.coef_[1]:+.3f} * musicStream")
print(f"       {lr.coef_[2]:+.3f} * onlineGames")

print("\n" + "-"*70)
print("INTERPRETATION:")
print("-"*70)
print(f"• Bias (intercept): {lr.intercept_:.3f}")
print(f"• Music Download: {lr.coef_[0]:.3f} (tiny positive effect)")
print(f"• Music Streaming: {lr.coef_[1]:.3f} (negative - reduces buying)")
print(f"• Online Games: {lr.coef_[2]:.3f} (strong positive - increases buying)")

print("\n" + "-"*70)
print("PROBLEM WITH LINEAR REGRESSION:")
print("-"*70)
print(f"• Predictions range from {y_pred_linear.min():.3f} to {y_pred_linear.max():.3f}")
print(f"• These are NOT probabilities (not bounded to [0,1])")
print(f"• Some predictions are negative or > 1")
```

```python
print("\n" + "="*70)
print("METHOD 2: LOGISTIC REGRESSION")
print("="*70)

# Train logistic regression
logreg = LogisticRegression(penalty=None, max_iter=10000, solver='lbfgs')
logreg.fit(X, y)

# Get probability predictions
y_pred_logistic_prob = logreg.predict_proba(X)[:, 1]
y_pred_logistic_class = logreg.predict(X)

# Create results table
results_logistic = pd.DataFrame({
    'Transaction': [f'T{i}' for i in range(1, 11)],
    'Actual': y,
    'Probability': y_pred_logistic_prob.round(3),
    'Predicted Class': y_pred_logistic_class
})

print("\nLogistic Regression Results:")
display(results_logistic)

# Display the learned function
print("\n" + "="*70)
print("LEARNED FUNCTION (Logistic Regression):")
print("="*70)
print(f"\nBuys = f(musicDown, musicStream, onlineGames)")
print(f"     = {logreg.intercept_[0]:.4f}")
print(f"       {logreg.coef_[0][0]:+.4f} * musicDown")
print(f"       {logreg.coef_[0][1]:+.4f} * musicStream")
print(f"       {logreg.coef_[0][2]:+.4f} * onlineGames")

print("\n" + "-"*70)
print("INTERPRETATION:")
print("-"*70)

coef_dict = {
    'Music Download': logreg.coef_[0][0],
    'Music Streaming': logreg.coef_[0][1],
    'Online Games': logreg.coef_[0][2]
}

print(f"• Bias (intercept): {logreg.intercept_[0]:.4f}\n")

for feature, coef in coef_dict.items():
    direction = "increases" if coef > 0 else "decreases"
    magnitude = abs(coef)
    strength = "strong" if magnitude > 10 else "moderate" if magnitude > 1 else "weak"
    
    print(f"• {feature}: {coef:.4f}")
    print(f"  → {direction} probability (coefficient is {coef:+.4f})")
    print(f"  → {strength} effect (|{magnitude:.4f}|)\n")

strongest_feature = max(coef_dict, key=lambda k: abs(coef_dict[k]))
print(f"Strongest predictor: {strongest_feature} (|{abs(coef_dict[strongest_feature]):.4f}|)")

print("\n" + "-"*70)
print("ADVANTAGES OF LOGISTIC REGRESSION:")
print("-"*70)
print(f"• All predictions are probabilities between 0 and 1")
print(f"• Range: {y_pred_logistic_prob.min():.3f} to {y_pred_logistic_prob.max():.3f}")
print(f"• Can interpret as probability of buying")
print(f"• Weights are much larger (more sensitive to features)")
```

```python
print("\n" + "="*70)
print("WHY ARE THE WEIGHTS DIFFERENT?")
print("="*70)

print("""
Linear Regression weights are SMALL:
    • Designed to predict continuous values near 0 or 1
    • Direct output: w^T·x

Logistic Regression weights are LARGE:
    • Need to be transformed through sigmoid: σ(w^T·x)
    • Large weights create steeper sigmoid curves
    • This gives more confident predictions (closer to 0 or 1)

The sigmoid function "squashes" large values into [0,1]:
    • σ(-35) ≈ 0.0
    • σ(0) = 0.5
    • σ(+35) ≈ 1.0
""")

# Visualize sigmoid with different weight magnitudes
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

z = np.linspace(-10, 10, 200)
sigmoid = lambda x: 1 / (1 + np.exp(-x))

# Different weight magnitudes
for scale in [0.5, 1, 2, 5]:
    ax.plot(z, sigmoid(scale * z), linewidth=2, label=f'Weight scale = {scale}')

ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)
ax.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
ax.set_xlabel('Raw score (w^T·x)', fontsize=13)
ax.set_ylabel('Probability σ(w^T·x)', fontsize=13)
ax.set_title('Effect of Weight Magnitude on Sigmoid', fontsize=14, weight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\nLarger weights → Steeper sigmoid → More confident predictions!")
```
